```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


def show_stats(df: pd.DataFrame, category: str):
    print(f"{category} descripton: ")
    print(df.groupby("website")["price"].describe())

    # Histogram → Shows price distributions across platforms.
    plt.figure(figsize=(10, 5))
    sns.histplot(data=df, x="price", hue="website", bins=30, kde=True)
    plt.title(f"Price Distribution of {category} Across Platforms")
    plt.suptitle(f"{category} Price Analysis Across Platforms", fontsize=14)
    plt.xlabel("Price (in USD)")
    plt.ylabel("Frequency")
    plt.show()

    # Bar Chart → Highlights average price differences.
    plt.figure(figsize=(8, 5))
    sns.barplot(x="website", y="price", data=df, estimator=np.mean)
    plt.title(f"Average {category} Prices Across Platforms")
    plt.suptitle(f"{category} Price Analysis Across Platforms", fontsize=14)
    plt.ylabel("Average Price")
    plt.show()

    # Line Chart → Tracks price trends over time.
    df["scraping_date"] = pd.to_datetime(df["scraping_date"])
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=df, x="scraping_date", y="price", hue="website")
    plt.title(f"Price Trends of {category} Over Time")
    plt.suptitle(f"{category} Price Analysis Across Platforms", fontsize=14)
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.xticks(rotation=45)
    plt.show()


show_stats(pd.read_csv("../cleaned_data/laptops.csv"), "laptops")
show_stats(pd.read_csv("../cleaned_data/smartphones.csv"), "smartphones")
show_stats(pd.read_csv("../cleaned_data/tablets.csv"), "tablets")
```
